```python
Df.head()
```

```python
C_list = list(Df['Country'].values)
```

```python
C_list_new= [x for x in C_list if type(x) != float ]
# print(C_list_new)
```

```python
from collections import defaultdict
C_dict = defaultdict(int)
```

```python
for i in C_list_new:
    
    c = i.split(';')
    
    for j in c:
        
        C_dict[j]+=1
```

```python
latlong_file = open('countries_LatLong_list.TXT', 'r')
latlong_list = latlong_file.readlines()
latlong_list = list(map(lambda x:x.strip().split('\t'), latlong_list))

country_names = {}

for i in latlong_list:
    country_names[i[0]]={'capital':'','lat':0,'long':0,'num_auth':0}
    country_names[i[0]]['capital'] = i[1]
    if i[2][-1] == 'S':
        country_names[i[0]]['lat'] = -1*float(i[2][:-1])
    else:
        country_names[i[0]]['lat'] = float(i[2][:-1])
    
    if i[3][-1] == 'W':
        country_names[i[0]]['long'] = -1*float(i[3][:-1])
    else:
        country_names[i[0]]['long'] = float(i[3][:-1])
```

```python
N = list(country_names.keys())
```

```python
for i in N:
    
    for j in C_dict:
        if i in j:
            country_names[i]['num_auth'] += C_dict[j]
```

```python
country_names['Australia'],C_dict['Australia']
```

```python
for j in C_dict:
    
    if 'USA' in j or 'U.S.A' in j or 'U S A' in 'j':
        
        country_names['United States of America']['num_auth'] += C_dict[j]

    if 'UK' in j or 'U.K' in j or 'United Kingdom' in 'j':
        
        country_names['United Kingdom of Great Britain and Northern Ireland']['num_auth'] += C_dict[j]
        
country_names['United States of America']
country_names['United Kingdom of Great Britain and Northern Ireland']
```

```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_nodes_from([c for c in list(country_names.keys()) if country_names[c]['num_auth']>0])
```
